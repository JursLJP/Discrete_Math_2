```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Adding nodes
G.add_nodes_from(range(6))  # Adding 6 nodes, labeled from 0 to 5

# Adding edges
edges = [(0, 1), (1, 2), (1, 3), (1, 4), (0, 2), (0, 3), (3, 3)]  # Including a self-loop at vertex 3
G.add_edges_from(edges)

# Drawing the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=700, font_size=15, font_color='darkblue', edge_color='gray')
plt.title('Example of a Simple Graph')
plt.show()
```

### Explanation of the Graph and the Handshake Theorem

This example illustrates the **Handshake Theorem** in graph theory, which states that the sum of the degrees of all vertices in a graph is twice the number of edges. This theorem is applicable in simple, undirected graphs.

#### Graph Composition
- **Vertices/Nodes**: \(A, B, C, D, E, F\).
- **Edges**: Connections that define interactions between vertices, including:
  - \(A - B\)
  - \(A - C\)
  - \(A - D\)
  - \(B - C\)
  - \(C - D\)
  - \(D - E\)
  - \(E - F\)
  - \(A - F\)

The degrees of the vertices are as follows:
\[
\begin{align*}
\text{Degree of } A & : 4 \\
\text{Degree of } B & : 2 \\
\text{Degree of } C & : 3 \\
\text{Degree of } D & : 3 \\
\text{Degree of } E & : 2 \\
\text{Degree of } F & : 2 \\
\end{align*}
\]

#### Handshake Theorem Validation
The sum of the degrees of all vertices is calculated, and according to the Handshake Theorem, it should be twice the number of edges in the graph:
\[
\text{Sum of degrees} = 4 + 2 + 3 + 3 + 2 + 2 = 16
\]
\[
\text{Twice the number of edges} = 2 \times 8 = 16
\]

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Adding nodes and edges
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'F'), ('A', 'F'), ('B', 'D'), ('A', 'C')]
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Finding a path from node A to node E
path = nx.shortest_path(G, source='A', target='E')

# Drawing the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_color='darkblue', edge_color='gray')

# Highlight the path from A to E
path_edges = list(zip(path, path[1:]))
nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='lightgreen')
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)
plt.title('Travelling in a Graph: Path from A to E')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Adding nodes and edges that form an Euler circuit
G.add_edges_from([('1', '2'), ('2', '3'), ('3', '4'), ('4', '5'), ('5', '1'), ('1', '3'), ('2', '4')])

# Check if the graph has an Eulerian Circuit
is_eulerian = nx.is_eulerian(G)
eulerian_circuit = list(nx.eulerian_circuit(G)) if is_eulerian else "No Eulerian circuit."

# Drawing the graph
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(G)  # Circular layout to evenly distribute nodes
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=16, font_color='darkblue', edge_color='gray')
plt.title('Graph with Euler\'s Circuit')
plt.show()

# Print Eulerian Circuit if it exists
print("Eulerian Circuit:", eulerian_circuit)
```

### Explanation of Euler's Circuit

Euler's Circuit is a path in a graph where each edge is used exactly once, and the path returns to the starting vertex. The graph shown meets the necessary conditions for Euler's Circuit, which are:

- **Connected Graph**: The graph is connected, meaning there is a path between every pair of vertices, which is essential for the existence of an Euler's circuit.
- **Positive Even Degree**: Every vertex in the graph has an even degree. This condition is crucial as it ensures that there are enough edges to enter and leave each vertex.

#### Conditions for Euler's Circuit:
- All vertices with nonzero degree are connected.
- All vertices have even degree.

#### Example of Euler's Circuit:
The graph includes vertices {1, 2, 3, 4, 5} and edges forming connections such as \(1-2-3-4-5-1\) with additional edges like \(1-3\) and \(2-4\) to maintain even degrees. An example Euler's Circuit could be \(1-2-4-3-1-5-4-2-3-5-1\), starting and ending at vertex 1 and traversing each edge exactly once.

### Visual Representation
The visualization provided demonstrates the graph structure with a circular layout, highlighting the connectivity and the ability to form an Euler's Circuit. Nodes are colored in light green, and edges are in gray, visually indicating the circuit's path if it is traced according to the Eulerian principles.

### Practical Implications
Euler's circuits are fundamental in applications requiring the traversal of every link or path without repetition, such as routing problems, network topology analysis, and in creating efficient tours in logistics and urban planning.



```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Adding nodes and edges that form potential for an Euler's Trail
G.add_edges_from([('1', '2'), ('2', '3'), ('3', '4'), ('4', '2'), ('2', '5')])

# Check if the graph has an Eulerian path
has_eulerian_path = nx.has_eulerian_path(G)
eulerian_path = list(nx.eulerian_path(G)) if has_eulerian_path else "No Eulerian path."

# Drawing the graph
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(G)  # Circular layout to evenly distribute nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_color='darkblue', edge_color='gray')
plt.title('Graph with Potential Euler\'s Trail')
plt.show()

# Print Eulerian Path if it exists
print("Eulerian Path:", eulerian_path)
```

### Explanation of Euler's Trail

Euler's Trail in a graph is a type of walk where each edge is used exactly once, starting at one vertex and ending at another. This trail passes through every vertex of the graph at least once, and can only occur if exactly two vertices have an odd degree, with all others having an even degree. The provided graph is used to demonstrate this concept.

#### Graph Configuration:
- **Vertices**: {1, 2, 3, 4, 5}
- **Edges**: {(1, 2), (2, 3), (3, 4), (4, 2), (2, 5)}

#### Conditions for Euler's Trail:
- The graph must be connected, ensuring a path exists between all pairs of vertices.
- Exactly two vertices must have an odd degree. In this example, vertices \(1\) and \(5\) have odd degrees (1 and 1, respectively), while all others have an even degree, making an Euler's Trail possible from \(1\) to \(5\).

#### Example of Euler's Trail:
An Euler's Trail can be described from \(v1\) to \(v5\), following the sequence: \(1 \to 2 \to 3 \to 4 \to 2 \to 5\). This trail covers all edges exactly once, demonstrating the necessary conditions for Euler's Trail.

### Visual Representation
The graph is displayed using a circular layout, with nodes color-coded in light blue and edges in gray, to clearly indicate the path of the trail. This visualization helps in understanding the traversal pattern necessary for an Euler's Trail.

### Practical Implications
Euler's Trails are important in solving problems where all paths must be visited once without retracing steps, applicable in routing, network design, and resource allocation problems.



```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph object
G = nx.Graph()

# Adding nodes and edges that form a Hamiltonian Circuit
G.add_edges_from([('1', '2'), ('2', '3'), ('3', '4'), ('4', '6'), ('6', '5'), ('5', '1')])

# Drawing the graph
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(G)  # Circular layout to evenly distribute nodes
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=16, font_color='darkblue', edge_color='gray')
plt.title('Graph with Hamiltonian Circuit')
plt.show()

# Attempt to find a Hamiltonian circuit (not built-in NetworkX, using simple check for demo)
def find_hamiltonian_path(G):
    all_paths = list(nx.simple_cycles(G))
    for path in all_paths:
        if len(path) == len(G.nodes()) and len(set(path)) == len(G.nodes()):
            return path + [path[0]]  # Return path making it a circuit
    return "No Hamiltonian circuit."

# Print Hamiltonian Circuit if it exists
hamiltonian_circuit = find_hamiltonian_path(G)
print("Hamiltonian Circuit:", hamiltonian_circuit)
```

### Explanation of Hamiltonian Circuit

A **Hamiltonian Circuit** is a loop in a graph that visits each vertex exactly once before returning to the starting vertex. This concept is vital in many applications, including scheduling, routing, and the traveling salesman problem. The provided graph demonstrates these properties.

#### Hamiltonian Circuit Conditions:
- **Contains Every Vertex**: The subgraph \(H\) includes every vertex of \(G\), as it involves a Hamiltonian circuit.
- **Connected**: \(H\) is connected, and the Hamiltonian circuit traverses through every vertex of the graph.
- **Same Number of Edges as Vertices**: Since the Hamiltonian circuit forms a cycle, the number of edges in \(H\) is equal to the number of vertices.
- **Degree-2 Property**: In the Hamiltonian circuit, every vertex, except the first and last (which are the same in a circuit), is incident to exactly two edges.

#### Example of a Hamiltonian Circuit:
An example of a Hamiltonian circuit in the graph could be: \(1 \to 2 \to 3 \to 4 \to 6 \to 5 \to 1\). This sequence visits each vertex exactly once and returns to the starting point, forming a complete loop.

### Visual Representation
The graph is visualized with a circular layout, highlighting the connectivity and the path of the Hamiltonian circuit. Nodes are colored in light blue and edges in gray to clearly demonstrate the circuit path.

### Practical Implications
Hamiltonian circuits are crucial in optimization problems where each node or location must be visited once without redundancy, optimizing travel or operational costs.



```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a weighted graph
G = nx.Graph()

# Adding nodes and weighted edges
edges = [('Home', 'Park', 5), ('Park', 'School', 6), ('School', 'Supermarket', 7),
         ('Supermarket', 'Work', 12), ('Work', 'Home', 15),
         ('Home', 'School', 7), ('Park', 'Supermarket', 9), ('Work', 'Park', 15)]
G.add_weighted_edges_from(edges)

# Ensure the graph is connected for demonstration purposes
for u in G.nodes():
    for v in G.nodes():
        if u != v and (u, v) not in G.edges:
            G.add_edge(u, v, weight=100)  # Adding a high weight to discourage use of non-direct paths

# Use the Christofides algorithm to find a solution to the TSP
from networkx.algorithms.approximation import traveling_salesman
tsp_path = traveling_salesman.christofides(G, weight='weight')

# Calculate total weight (travel time) of the path
total_time = sum(G[u][v]['weight'] for u, v in zip(tsp_path[:-1], tsp_path[1:]))

# Drawing the graph with the path
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=16)
nx.draw_networkx_edges(G, pos, edgelist=list(zip(tsp_path[:-1], tsp_path[1:])), width=2, edge_color='red')
plt.title('Traveling Salesman Problem Approximation')
plt.show()

# Output the path and total travel time
print("TSP Path:", tsp_path)
print("Total Travel Time:", total_time, "mins")
```

### Explanation of the Traveling Salesman Problem (TSP)

The Traveling Salesman Problem (TSP) is a classic optimization challenge that involves finding the shortest possible route that visits each location exactly once and returns to the origin point. It's pivotal in logistics and operations management.

#### Scenario:
The graph represents various locations with travel times between them as weighted edges. The objective is to minimize the total travel time while visiting each location.

#### Proposed Routes and Their Total Times:
- **Route 1**: Home -> Park -> School -> Supermarket -> Work -> Home (Total time: 65 mins)
- **Route 2**: Home -> School -> Park -> Supermarket -> Work -> Home (Total time: 62 mins)
- **Route 3** and others, each detailing the path and total time.

The optimal path determined by a simple heuristic may not always be the shortest due to the complexity of the problem, but it provides a good approximation.

### Solution and Visual Representation
The solution path is highlighted, with nodes color-coded in light blue and the path in red. This visual aid helps to comprehend the route taken and the efficiency of the approximation algorithm used.

### Practical Implications
The TSP has profound implications in route planning for deliveries, scheduling, and even in circuit design, where minimizing the path is crucial for efficiency and cost reduction.
