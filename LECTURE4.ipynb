### Kruskal's Algorithm Implementation


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])  # Including 'I' to maintain completeness

# Add weighted edges
G.add_weighted_edges_from([
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
])

# Generate a planar layout
pos = nx.planar_layout(G)  # Use planar layout for a flat, non-overlapping visual

# Drawing the entire graph with a pink background for nodes
plt.figure(figsize=(10, 8))  # Setting the size of the figure
nx.draw_networkx(G, pos, with_labels=True, node_color='pink', edge_color='grey', node_size=500)

# Adding edge labels with red font to highlight weights
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Kruskal's algorithm to determine the minimum spanning tree
mst = nx.minimum_spanning_tree(G, algorithm='kruskal')
mst_edges = list(mst.edges())

# Highlight the MST edges in red and thicker
nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)

plt.title('Graph with Kruskal\'s Minimum Spanning Tree Implemented')
plt.axis('off')  # This will turn off the axis which removes the box around the graph
plt.show()
```

### Kruskal's Algorithm Implementation Explained

**Input:** G (a connected weighted graph with n vertices, where n is a positive integer)

**Algorithm Steps:**
1. **Initialize the Subgraph T:** Begin with a subgraph T that contains all vertices of G but no edges.
2. **Edge Set and Initial Edge Count:** Let E be the set of all edges of G, and initialize an edge count m = 0.
3. **Constructing the Minimum Spanning Tree (MST):**
   - **While Loop:** Continue until the number of edges in T is less than n - 1 (m < n - 1):
     - **Find Minimum Weight Edge:** Identify and select an edge e from E that has the least weight.
     - **Delete the Edge:** Remove this edge e from the set E to avoid reconsideration.
     - **Check and Add Edge:** Add edge e to T if it does not form a cycle with the already existing edges in T. If it is added, increment m by 1.

**Output:** The subgraph T, which is a minimum spanning tree for G, containing all vertices connected by the edges added in order of their increasing weight, ensuring minimal total edge weight without any cycles.


### Prim's Algorithm Implementation

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])  # Include node 'I'

# Add weighted edges
G.add_weighted_edges_from([
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
])

# Generate a planar layout for consistent node placement
pos = nx.planar_layout(G)

# Applying Prim's Algorithm to find the minimum spanning tree
mst = nx.minimum_spanning_tree(G, algorithm='prim')
mst_edges = list(mst.edges())

# Drawing the graph
plt.figure(figsize=(10, 8))
nx.draw_networkx(G, pos, with_labels=True, node_color='pink', edge_color='grey', node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=mst_edges, edge_color='red', width=2)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title('Graph with Prim\'s Minimum Spanning Tree Highlighted')
plt.axis('off')  # Remove the axes
plt.show()
```

### Prim's Algorithm Implementation Explained

**Input:** G (a connected weighted graph with n vertices, where n is a positive integer)

**Algorithm Steps:**
1. **Initialize the Spanning Tree (ST):** Start with a tree T containing a single vertex (typically the first or any specified vertex).
2. **Edge Priority Queue:** Maintain a priority queue that will hold all edges crossing the cut, initialized with all edges connected to the starting vertex.
3. **Expand the Spanning Tree:**
   - **While Loop:** Continue until T includes all vertices of G:
     - **Extract Minimum Weight Edge:** Remove and get the edge e with the lowest weight from the priority queue that connects a vertex in T to a vertex not in T.
     - **Add Edge and Vertex to T:** Add this edge e and its opposite vertex to T.
     - **Update the Priority Queue:** For the newly added vertex, insert all its connecting edges that connect to vertices not yet in T into the priority queue.

**Output:** The tree T is the minimum spanning tree of G, which connects all vertices with the minimum total edge weight without forming any cycles.


### Dijkstra's Algorithm Implementation

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G", "I"])  # Include node 'I'

# Add weighted edges
G.add_weighted_edges_from([
    ("A", "C", 10), ("A", "I", 5), ("C", "B", 20), ("C", "E", 15),
    ("C", "D", 30), ("B", "E", 25), ("I", "E", 40), ("E", "F", 35),
    ("F", "G", 50), ("G", "E", 45)
])

# Applying Dijkstra's Algorithm from node 'A'
lengths, paths = nx.single_source_dijkstra(G, source='A')

# Draw the graph
pos = nx.planar_layout(G)  # Consistent node placement
plt.figure(figsize=(10, 8))
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', edge_color='grey', node_size=500)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Highlight the shortest path tree
for path in paths.values():
    nx.draw_networkx_edges(G, pos, edgelist=[(path[i], path[i + 1]) for i in range(len(path) - 1)], edge_color='red', width=2)

plt.title('Graph with Dijkstra\'s Shortest Path Tree Implementation')
plt.axis('off')  # Remove the axes
plt.show()
```

### Dijkstra's Algorithm for Shortest Paths Explained

**Input:** G (a connected weighted graph with vertices, where each edge has a non-negative weight)

**Algorithm Steps:**
1. **Initialize Distance and Path Structures:** Set the distance to the start vertex as zero and to infinity for all other vertices. Maintain a priority queue of all vertices based on their current known distance.
2. **Set Starting Point:** Begin with the starting vertex, typically denoted as source.
3. **Algorithm Execution:**
   - **While Loop:** While the priority queue is not empty:
     - **Vertex with Minimum Distance:** Remove the vertex u from the priority queue which has the smallest distance.
     - **Relaxation:** For each neighbor v of u that is still in the priority queue, if the distance to u plus the edge weight from u to v is less than the known distance to v, update the distance to v and update vs predecessor to u.
     
**Output:** The distances to each vertex from the start vertex represent the shortest paths, and the predecessors can reconstruct the actual path taken.

**Visualization:** The shortest paths from the source are highlighted, demonstrating the efficiency of Dijkstra's approach in navigating graphs.
