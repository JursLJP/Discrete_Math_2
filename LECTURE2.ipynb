## Task 1: Question 1

## Expected Value of \(Y_i\)

### Description of the Experiment

1. **Number of Trials**: The experiment consists of \(n\) identical trials.
2. **Outcome Classification**: The outcome of each trial falls into one of \(k\) classes or cells.
3. **Probability**: The probability that the outcome of a trial falls into cell \(i\) is \(p_i\), with \(\sum_{i=1}^k p_i = 1\).
4. **Independence**: The trials are independent.

### Variables of Interest

- **Random Variables**: \(Y_1, Y_2, \ldots, Y_k\), where \(Y_i\) equals the number of trials for which the outcome falls into cell \(i\).
- **Sum of Outcomes**: \(Y_1 + Y_2 + \ldots + Y_k = n\).

### Joint Probability Distribution

The joint probability distribution of \(Y\) is given by:
\[
P(y) = \frac{n!}{y_1! y_2! \ldots y_k!} p_1^{y_1} p_2^{y_2} \ldots p_k^{y_k}
\]

### Task: Show the Expected Value

We need to show:
\[
E[Y] = \sum_y Y \cdot P(y) = n \cdot p_i
\]

### Proof Using the Properties of Expectation

For a multinomial distribution, the expectation of \(Y_i\) is given by:
\[
E[Y_i] = n \cdot p_i
\]

#### Reasoning

- **Multinomial Setup**: Each trial results independently in one of \(k\) outcomes with probabilities \(p_1, p_2, \ldots, p_k\).
- **Expectation of Binomial Trials**: Each \(Y_i\) behaves as if it were binomially distributed, since \(Y_i\) counts the number of times the outcome \(i\) occurs over \(n\) trials.
- **Binomial Expectation**: The expected value of a binomial random variable \(Y_i\) with parameters \(n\) and \(p_i\) (number of trials and success probability, respectively) is \(E[Y_i] = n \cdot p_i\).

### Conclusion

Given the setup and properties of multinomial distributions, it is shown that:
\[
E[Y_i] = n \cdot p_i
\]
This confirms that the expected number of times an outcome falls into a specific class \(i\) over \(n\) trials is indeed \(n \cdot p_i\).

### Problem Statement

If 500 adults are sampled randomly, calculate:

   - **100 persons between 18 and 24**,
   - **200 persons between 25 and 34**,
   - **200 persons between 45 and 64**.

### Task 1: Probability Calculation Using Multinomial Distribution

Given that the sampling process is random and each sample falls independently into these categories according to the given proportions, we use the multinomial distribution. The probability mass function for a multinomial distribution is given by:

\[
P(y_1, y_2, y_3, y_4, y_5) = \frac{n!}{y_1! y_2! y_3! y_4! y_5!} p_1^{y_1} p_2^{y_2} p_3^{y_3} p_4^{y_4} p_5^{y_5}
\]

where \(n = 500\), \(p_1 = 0.18\), \(p_2 = 0.23\), \(p_3 = 0.16\), \(p_4 = 0.27\), and \(p_5 = 0.16\). The specific probabilities for the scenario described are:

\[
P(100, 200, 0, 200, 0) = \frac{500!}{100! \cdot 200! \cdot 0! \cdot 200! \cdot 0!} \cdot 0.18^{100} \cdot 0.23^{200} \cdot 0.16^0 \cdot 0.27^{200} \cdot 0.16^0
\]

### Task 2: Expected Value Calculation

The expected number of people in the age group 65 and above, when 500 adults are sampled, is calculated as:
\[
E[Y_5] = n \cdot p_5 = 500 \cdot 0.16 = 80
\]

### Conclusion

1. The calculated probability using the multinomial distribution formula provides the likelihood of obtaining exactly 100 persons aged 18-24, 200 persons aged 25-34, and 200 persons aged 45-64 in a random sample of 500 people.
2. The expected number of people aged 65 and above in such a sample is 80.


## Task 2: Question 1

## Deriving the Probability Distribution

### Experiment Description

The experiment is described as follows:
1. **Number of Trials**: A fixed number, \(n\), of identical trials.
2. **Outcomes**: Each trial has two possible outcomes:
   - **Success** (\(S\))
   - **Failure** (\(F\))
3. **Probability of Success**: The probability of success on a single trial is \(p\).
4. **Probability of Failure**: The probability of failure is \(q = 1 - p\).
5. **Independence**: All trials are independent.

### Random Variable of Interest

- **\(Y\)**: The random variable \(Y\) represents the number of successes observed during the \(n\) trials.

### Derivation of the Probability Distribution

Given the setup, \(Y\) follows a binomial distribution because it counts the number of successes in \(n\) independent trials, each with the same probability of success \(p\). The probability distribution of \(Y\) can be derived as follows:

#### Probability Mass Function (PMF)

The probability mass function of a binomial random variable \(Y\) is given by:
\[
P(Y = k) = \binom{n}{k} p^k (1-p)^{n-k}
\]
for \(k = 0, 1, 2, \ldots, n\), where:
- \(\binom{n}{k}\) is the binomial coefficient, representing the number of ways to choose \(k\) successes out of \(n\) trials.

### Formula Explanation

- **Binomial Coefficient**: \(\binom{n}{k} = \frac{n!}{k!(n-k)!}\) counts the number of ways to arrange \(k\) successes within \(n\) trials.
- **\(p^k\)**: The probability of having exactly \(k\) successes.
- **\((1-p)^{n-k}\)**: The probability of the remaining \(n-k\) trials resulting in failure.

### Conclusion

The random variable \(Y\), representing the number of successes in \(n\) trials of this experiment, follows a binomial distribution with parameters \(n\) (number of trials) and \(p\) (probability of success). The derived PMF accurately reflects the likelihood of observing any specific number of successes.

## Task 2, Question 2: Expected Value of the Binomial Distribution

### Expectation Formula for Binomial Distribution

Given the binomial distribution derived in the previous question, we calculate the expected value (mean) of \(Y\), which represents the number of successes in \(n\) trials.

### Expected Value Calculation

The expected value \(E[Y]\) of a binomially distributed random variable \(Y\) with parameters \(n\) (number of trials) and \(p\) (probability of success) is given by:
\[
E[Y] = n \cdot p
\]

#### Derivation

This result can be derived from the definition of expectation:
\[
E[Y] = \sum_{k=0}^n k \cdot P(Y = k) = \sum_{k=0}^n k \cdot \binom{n}{k} p^k (1-p)^{n-k}
\]

Using the properties of binomial coefficients and the simplification of the sum, we obtain the expected value formula for a binomial distribution.

### Conclusion

The expected number of successes \(E[Y]\) in \(n\) trials, each with a probability \(p\) of success, is exactly \(n \cdot p\). This confirms the consistency of the binomial distribution's properties with elementary probability theory.


## Task 2, Question 3: Probability of Recovery Without Effective Medication

### Background

Experience indicates that 30% of all persons afflicted by a certain illness recover naturally. Assuming a medication is ineffective (placebo), we consider the scenario where 10 people receive this medication.

### Problem Statement

Calculate the probability that at least nine out of these ten people recover.

### Calculation Using Binomial Distribution

Given:
- \(n = 10\) (the number of people),
- \(p = 0.30\) (the natural recovery rate without medication).

The variable of interest is \(Y\), the number of recoveries, which follows a binomial distribution \(Y \sim \text{Binomial}(10, 0.30)\).

#### Probability Mass Function (PMF)

The PMF for \(Y\) is:
\[
P(Y = k) = \binom{10}{k} \cdot 0.30^k \cdot 0.70^{10-k}
\]

#### Probability of at Least Nine Recoveries

We need to calculate:
\[
P(Y \geq 9) = P(Y = 9) + P(Y = 10)
\]

\[
P(Y = 9) = \binom{10}{9} \cdot 0.30^9 \cdot 0.70^{1} 
\]
\[
P(Y = 10) = \binom{10}{10} \cdot 0.30^{10} \cdot 0.70^{0}
\]

### Conclusion

The total probability of at least nine people recovering, under the assumption that the medication is ineffective, is the sum of the probabilities for \(Y = 9\) and \(Y = 10\). This probability quantifies the chance of observing a high number of recoveries due to natural healing processes rather than the effect of the medication.


## Task 2, Question 4: Probability of Detecting a Defective Fuse

### Problem Statement

Given a lot of 5000 electrical fuses, where 5% are defective, we need to find the probability of observing at least one defective fuse in a sample of five fuses.

### Approach

Since the lot size is relatively large compared to the sample size, we can use the binomial probability model to approximate the probability of defects.

### Probability Setup

- **\(n = 5\)**: The sample size (number of fuses tested).
- **\(p = 0.05\)**: Probability of finding a defective fuse.

The variable of interest is \(X\), the number of defective fuses found in the sample, which follows a binomial distribution:
\[
X \sim \text{Binomial}(n, p)
\]

### Calculating the Probability

The probability of finding at least one defective in the sample is:
\[
P(X \geq 1) = 1 - P(X = 0)
\]
\[
P(X = 0) = \binom{5}{0} \cdot 0.05^0 \cdot 0.95^5 = 0.95^5
\]
\[
P(X \geq 1) = 1 - 0.95^5
\]

### Computation

Using the formula for binomial probabilities:
\[
P(X \geq 1) = 1 - (0.95)^5
\]

This calculation gives us the probability that in a randomly selected sample of five fuses, at least one will be defective.

### Conclusion

The result shows how likely it is to encounter a defective fuse in small sample testing, given the defect rate in the larger lot. This information is crucial for quality control in manufacturing processes.


## Task 3, Question 1: Expected Value of Poisson Distribution

### Poisson Distribution Formula

The Poisson distribution is given by:
\[
p(y) = \frac{\lambda^y e^{-\lambda}}{y!}
\]
where \(\lambda\) is the average number of occurrences in an interval and \(y\) is the number of occurrences.

### Expected Value Calculation

The expected value \(E[Y]\) of a Poisson distribution is one of its well-known properties. The formula for the expected value of the Poisson distribution is simply:
\[
E[Y] = \lambda
\]

#### Reasoning

This result arises from the definition of expectation for discrete random variables:
\[
E[Y] = \sum_{y=0}^\infty y \cdot p(y) = \sum_{y=0}^\infty y \cdot \frac{\lambda^y e^{-\lambda}}{y!}
\]

By properties of the exponential function and the series expansion, this simplifies directly to:
\[
E[Y] = \lambda
\]

### Conclusion

The expected value of a Poisson-distributed random variable represents the mean number of events in the given interval, which is \(\lambda\).


## Task 3, Question 2: Probability of No Seedlings in Sampling Regions

### Scenario Description

Given that seedlings are randomly dispersed with a mean density of approximately five per square yard, we need to calculate the probability that none of the ten sampled 1-square-yard regions will contain seedlings.

### Probability Setup Using Poisson Distribution

The random variable \(Y\), representing the number of seedlings per square yard, follows a Poisson distribution with \(\lambda = 5\) (the mean density of seedlings per square yard).

### Calculating the Probability

The probability that a single 1-square-yard region contains no seedlings is:
\[
P(Y = 0) = \frac{5^0 e^{-5}}{0!} = e^{-5}
\]

#### Probability for Ten Regions

Since the regions are sampled independently, the probability that none of ten regions contain seedlings is:
\[
P(\text{None of 10 regions contain seedlings}) = (e^{-5})^{10} = e^{-50}
\]

### Conclusion

The probability that none of ten randomly selected 1-square-yard regions will contain seedlings is \(e^{-50}\). This very low probability reflects the high expected density of seedlings per region.
